```python
q1=df.quantile(0.25)
q2=df.quantile(0.75)
IQR = q2-q1
IQR
```

```python
df = df[~((df < (q1 - 1.5 * IQR)) |(df > (q2 + 1.5 * IQR))).any(axis=1)]
df.shape
```
